```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
url="/content/wordcloud.csv"
```

```python
df=pd.read_csv(url,encoding="latin-1")
```

```python
df
```

```python
comment_words=" "
```

```python

stp=set(STOPWORDS)
```

```python
stp
```

```python
for val in df.name:
  val=str(val)

  tokens=val.split()

  for i in range(len(tokens)):
    tokens[i]=tokens[i].lower()

  comment_words+="  ".join(tokens)+"  "
```

```python

wordcloud=WordCloud(background_color="tan",repeat=True, mask=None,width=1586,random_state=175
                    ,relative_scaling='auto', height=396).generate(comment_words)
```

```python

plt.imshow(wordcloud)
plt.axis("off")
```
